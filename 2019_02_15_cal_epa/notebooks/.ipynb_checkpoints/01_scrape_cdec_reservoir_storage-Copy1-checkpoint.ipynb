# Scrape CDEC reservoir storage data

CDEC gives reservior storage data at a temporal resolution as fine as the hour. Let's grab some of that sweet data from the past month. 

This is the website we're going to construct queries for and scrape: https://cdec.water.ca.gov/dynamicapp/wsSensorData

but we're just going to grab their monthly reservoir storage product, which is all we need. Let's grab all the data from the 1974 water year ($t_0$ for C2VSimFG beta) through today!

```r
# packages
library(tidyverse) # general purpose DS toolkit
library(rvest)     # for webscraping
library(viridis)   # colorblind-safe color palettes
library(readr)     # fast read/write functions
```

```r
# scrape CDEC station codes
station_codes <- read_html('http://cdec.water.ca.gov/misc/resinfo.html') %>% 
  html_table() %>% 
  .[[1]] %>% 
  pull(ID)
head(station_codes)
```

```r
# starting and ending time indices
start <- '2019-01-01'               # start 
end   <- '2019-02-14'               # today

# folsom and shasta stations
fol_sha <- c("FOL", "SHA")

# constrct the url to query CDEC
url <- paste0('https://cdec.water.ca.gov/dynamicapp/req/CSVDataServlet?Stations=', 
       paste(fol_sha, collapse = '%2C'),
       '&SensorNums=45&dur_code=D&Start=', # daily precip in inches
       #'&SensorNums=15&dur_code=D&Start=', # daily storage in AF
       start,
       '&End=',
       end)

# read the generated webpage, clean the date column, and omit missing values for the present day
df <- read.csv(url)

df <- df %>% 
  mutate(year = substr(df$OBS.DATE, 1, 4), 
         month = substr(df$OBS.DATE, 5, 6), 
         day = substr(df$OBS.DATE, 7, 8), 
         ymd = lubridate::ymd(paste(year, month, day, sep = "-"))) %>% 
  filter(VALUE != "---") %>% 
  mutate(VALUE = as.numeric(VALUE))
```

```r
# view what we've got
df
```

***  

## Visualize  

There are 199 station codes, but we're just looking at the storage in 2 reservoirs over the past month: Folsom and Shasta.

```r
df %>%
  ggplot(aes(ymd, VALUE, color = STATION_ID)) + 
  geom_line() + 
  scale_color_viridis_d() +
  labs(title = "Precipitation at FOL and SHA",
       x = "Date", y = "Precip (inches)")
```
